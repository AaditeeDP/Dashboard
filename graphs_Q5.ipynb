```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```
df_original = pd.read_csv('Q5_2.csv')
df = df_original.copy()
def get_top_values(df):
    return df['Company Name'].value_counts().idxmax() 

topdf = df.groupby('Orders').apply(get_top_values) 
```

```
topdf.tail(30)
```

```
topdf.to_csv(r'Q5_orders.csv')
```

```
##making stack bar chart
orders = pd.read_csv('top_50_orders.csv')
fig = px.bar(orders, x="Company Name", y="Orders", title="Most ordering companies")
fig.show()
```

```
units = pd.read_csv('top_50_units.csv')
fig = px.bar(units, x="Company Name", y="Units", title="Most units ordering companies")
fig.update_xaxes(categoryorder='total descending')
fig.show()
```
